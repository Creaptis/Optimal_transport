```python
def FLB(C1, C2, p, q) : # we dropped the 1/2 factor and the 1/p power
    S_x = (np.sum(C1**2*p.reshape(-1,1) , axis = 0)**0.5).reshape(-1,1)
    S_y = (np.sum(C2**2*q.reshape(-1,1) , axis = 0)**0.5).reshape(-1,1)
    
    G , log = ot.lp.solver_1d.emd_1d(S_x, S_y, a=p, b=q, metric='sqeuclidean', p=2., dense=True, log=True)
    
    return(G, log['cost'])
    
```

```python
def SLB(C1, C2, p, q) :
    P_2 = np.outer(p,p).flatten()
    C1_2 = C1.flatten()
    Q_2 = np.outer(q,q).flatten()
    C2_2 = C2.flatten()
    Cost_Matrix = np.zeros((len(C1_2), len(C2_2)) )
    
#    for i in range(Cost_Matrix.shape[0]) :
#        for j in range(Cost_Matrix.shape[1]) :
#            Cost_Matrix[i][j] = (C1_2[i] - C2_2[j] )**2
    
    Cost_Matrix = (C1_2[:,None] - C2_2[None,:])**2
    
    
    G, log = ot.lp.emd(a= P_2, b=Q_2, M=Cost_Matrix, numItermax=100000, log=True, center_dual=True, numThreads=1)
    
#    print("RRRRRRRRRR")
#    cost = np.sum(G*Cost_Matrix)
#    print(cost)
    cost = log['cost']
    return(G, cost)
```

```python
def TLB(C1, C2, p, q) :
    
    OMEGA = np.zeros((len(p), len(q)) )
    
    for i in range(OMEGA.shape[0]):
        for j in range(OMEGA.shape[1]) :
            x_a = C1[i]
            x_b = C2[j]
            gamma , log = ot.lp.solver_1d.emd_1d(x_a, x_b, a=p, b=q, metric='sqeuclidean', p=2., dense=True, log=True)
            
#            print('qsegqsegf')
#            print(log['cost'])
#            print(np.sum( gamma*(x_a[:,None] - x_b[None,:])**2 ))
            
            
            OMEGA[i][j] = log['cost']
    G, log = ot.lp.emd(a  = p, b = q, M = OMEGA, numItermax=100000, log=True, center_dual=True, numThreads=1)
    cost = log['cost']
    return(G, cost)    
```

```python
def generate_ortho_matrix(dimension) :
    RESULT = np.zeros((dimension, dimension)).astype("float64")
    RESULT[:,0] = np.random.rand(dimension)
    RESULT[:,0] /= np.linalg.norm(RESULT[:,0])
    for i in range(1,dimension) :
        RESULT[:,i] = np.random.rand(dimension)
        for j in range(i) :
            RESULT[:,i] -= np.dot(RESULT[:,i],RESULT[:,j])*RESULT[:,j]
        RESULT[:,i] /= np.linalg.norm(RESULT[:,i])
    return(RESULT)

def random_isometrie_nD(X) :
    X = X[:,:,None]
    
    MAT = generate_ortho_matrix(dimension)
    
    X = (MAT@X).reshape(-1,dimension)
    print(X.shape)
    return(X)

def random_shuffle(xs, p) :
    shuffle = list(range(len(xs)))
    np.random.shuffle(shuffle)
    xt = xs[shuffle ]
    q = p[shuffle]
    return(xt,q)


def check_if_valid(p,q,T) :
    q_hat = np.sum(T , axis = 0)
    p_hat = np.sum(T , axis = 1)
    epsilon = 1e-10
    
    return(  ( q_hat - q - epsilon < 0 ).all() and ( q_hat - q + epsilon > 0 ).all() and ( p_hat - p - epsilon < 0 ).all() and ( p_hat - p + epsilon > 0 ).all()  )
    

def random_initialisation(p,q, nb_itermax = 1000) :
    # returns a random valid transport between distributions of respective histogram p and q
    
    T = np.random.rand(len(p),len(q)).astype('float64')
    T_new = copy.deepcopy(T) + 1 
    count = 0
    while count < nb_itermax and not check_if_valid(p,q,T) :
        T = copy.deepcopy(T_new)
        T_new = T*q.reshape(1,-1)/ np.sum(T, axis = 0).reshape(1,-1)
        T_new = T_new*p.reshape(-1,1)/ np.sum(T_new, axis = 1).reshape(-1,1)
        count+=1
        
    if count == nb_itermax :
        print("ATTENTION, la précision voulue n'a pas été atteinte")
    return(T_new)
```
